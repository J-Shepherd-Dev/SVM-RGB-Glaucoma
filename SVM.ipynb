```python
import os
import glob
import numpy as np
import pandas as pd
from PIL import Image
from sklearn import svm
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
import pickle

# Set path to image folder
image_folder_path = r'C:\Users\johns\OneDrive\University work\Final Year\Final Year Project\Datasets\archive (1)\ORIGA\ORIGA\Flatten images'

# Create flattened directory
flattened_folder_path = os.path.join(image_folder_path, 'flattened 2')
os.makedirs(flattened_folder_path, exist_ok=True)

# Get list of all image files in folder
image_files = glob.glob(os.path.join(image_folder_path, '*.jpg'))

# Define image size
image_size = (512, 512)

# Load flattened images into an array
data = []
for image_file in image_files:
    # Load image and resize
    image = Image.open(image_file).resize(image_size)
    flattened_array = np.array(image).flatten()
    data.append(flattened_array)

    # Reshape flattened array to (512, 512, 3) and save image
    reshaped_array = flattened_array.reshape((512, 512, 3))
    final_image = Image.fromarray(np.uint8(reshaped_array))

    flattened_image_file = os.path.join(flattened_folder_path, os.path.basename(image_file))
    final_image.save(flattened_image_file)

# Load spreadsheet data and map to image filenames
image_labels = df["Glaucoma"]

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(data, image_labels, test_size=0.3, random_state=42)

# Train SVM model
clf = svm.SVC(kernel='linear')
clf.fit(X_train, y_train)
#test cross validation score
scores = cross_val_score(clf, X_train, y_train, cv=5)
print(scores)
# Test SVM model
accuracy = clf.score(X_test, y_test)
print(f"Accuracy: {accuracy}")



filename = 'RGB_SVM_model.sav'
pickle.dump(clf, open(filename, 'wb'))
```

```python
import os
import glob
import numpy as np
from PIL import Image
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from tensorflow.keras.applications.resnet50 import ResNet50, preprocess_input

# Set path to image folder
image_folder_path = r'C:\Users\johns\OneDrive\University work\Final Year\Final Year Project\Datasets\archive (1)\ORIGA\ORIGA\Flatten images\Flattened 2'

# Get list of all image files in folder
image_files = glob.glob(os.path.join(image_folder_path, '*.jpg'))

# Define image size
image_size = (512, 512)

# Load flattened images into an array
data = []
for image_file in image_files:
    # Load image and resize
    image = Image.open(image_file).resize(image_size)
    flattened_array = np.array(image).reshape(-1)
    data.append(flattened_array)

# Load spreadsheet data and map to image filenames
image_labels = df["Glaucoma"]

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(data, image_labels, test_size=0.3, random_state=42)

# Load ResNet50 model
resnet_model = ResNet50(weights='imagenet', include_top=False, input_shape=(image_size[0], image_size[1], 3))

# Extract features from images
X_train_features = resnet_model.predict(np.array(X_train).reshape(-1, image_size[0], image_size[1], 3))
X_train_features = X_train_features.reshape(X_train_features.shape[0], -1)

X_test_features = resnet_model.predict(np.array(X_test).reshape(-1, image_size[0], image_size[1], 3))
X_test_features = X_test_features.reshape(X_test_features.shape[0], -1)

# Train SVM on features
svm_model = LinearSVC()
svm_model.fit(X_train_features, y_train)

# Make predictions on test set
y_pred = svm_model.predict(X_test_features)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print('Accuracy:', accuracy)

# Extract features from all images
data_features = resnet_model.predict(np.array(data).reshape(-1, image_size[0], image_size[1], 3))
data_features = data_features.reshape(data_features.shape[0], -1)

# Make predictions on all data
all_predictions = svm_model.predict(data_features)

#cm = confusion_matrix(image_labels, all_predictions)
#print('Confusion Matrix:')
#print(cm)

# Calculate confusion matrix
predictions = all_predictions
TN, FP, FN, TP = confusion_matrix(image_labels, predictions).ravel()

print('True Positive(TP)  = ', TP)
print('False Positive(FP) = ', FP)
print('True Negative(TN)  = ', TN)
print('False Negative(FN) = ', FN)

#Classification report
print('Classification Report:')
print(classification_report(image_labels, all_predictions))
#test cross validation score
scores = cross_val_score(svm_model, X_train, y_train, cv=5)
print(scores)

# Save SVM model
import pickle
with open('RGB_SVM_model2.sav', 'wb') as f:
    pickle.dump(svm_model, f)
```
